#Lemmatization

Cette fonction permet de remplacer les mots pluriels en singulier dans une chaine de caractère

```python
def lemmatization(sentence) :
  p = inflect.engine()
  mots = nltk.word_tokenize(sentence)
  mots_singulier = []
  for mot in mots:
    mot_singulier = p.singular_noun(mot)
    if mot_singulier:
        mots_singulier.append(mot_singulier)
    else:
        mots_singulier.append(mot)
  phrase_singulier = ' '.join(mots_singulier)
  return(phrase_singulier)


```

On applique la fonction lemmatize aux données dans la colonne Libellé produit

```python
for i in range (len(df2)) :
  element = df2.iloc[i, 0]
  singulier = lemmatization(element)
  df2.loc[i, "Libellé produit"] = singulier
```

Cette fonction permet de trouver la catégorie de inputString sinon elle retourne none

```python


def findCategory2(inputString, nature_unique2 = nature_unique3):
    inputString = inputString.lower()
    matchedCategory = None
    for i in range(nature_unique2.shape[0]):
        category = nature_unique2['Nature'][i].lower()
        if re.search(r'\b' + re.escape(category) + r'\b', inputString):
            matchedCategory = nature_unique2['Nature'][i]
            break
    return matchedCategory
```
